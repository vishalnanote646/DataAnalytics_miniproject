```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
import pandas as pd
```

```python
df=pd.read_csv('dataset.csv')
```

```python
df.rename(columns={
    'player': 'batsman_name',
    'team': 'country',
    'opponent': 'opposition_team',
    'strike_rate': 'SR',
    'innings': 'batting_innings'
}, inplace=True)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df.groupby('country')['runs'].sum().reset_index(), x='country', y='runs', palette='Blues_d')
plt.title("Total Runs by Country")
plt.ylabel("Runs")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['runs'], marker='o', linestyle='-', color='green')
plt.title("Runs Over Player Index")
plt.xlabel("Index")
plt.ylabel("Runs")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['balls'], bins=10, color='orange', edgecolor='black')
plt.title("Distribution of Balls Faced")
plt.xlabel("Balls")
plt.ylabel("Frequency")
plt.show()
```

```python
country_counts = df['country'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Players per Country")
plt.axis('equal')  # Makes it a circle
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='runs', y='SR', hue='country')
plt.title("Runs vs Strike Rate")
plt.xlabel("Runs")
plt.ylabel("Strike Rate")
plt.legend(title='Country')
plt.show()
```
